```python
from fem.core.Node         import Node
from fem.core.Material     import Material
from fem.sections.Membrane import Membrane

from fem.elements.CST      import CST
from fem.elements.LST      import LST
from fem.elements.Quad4    import Quad4
from fem.elements.Quad9    import Quad9

import numpy as np
```

```python
# Material
Steel = Material(name='Steel', E=0.0, nu=0.3, rho=0.0 , printSummary=True)
F1    = Membrane(name='Plate', thickness=1.0, material=Steel)
```

```python
# Monkey patch directo
Steel.get_Emat = lambda type=None: np.array([
    [4, 1, 0],
    [1, 4, 0],
    [0, 0, 2],
]) * 8

# Verificar
print(Steel.get_Emat())
```

```python
# Nodes
n1 = Node(0, [0.0, 0.0])
n2 = Node(1, [3.0, 1.0])
n3 = Node(2, [2.0, 2.0])
```

```python
# Element
e1 = CST(element_tag=1, 
        node_list=[n1, n2, n3], 
        section=F1)
```

```python
# Outputs
print("Area:")
print(e1.area)

print("\nB matrix:")
print(e1.get_B_matrix())

print("\nN at centroid:")
centroid = e1.get_centroid()
print("Centroid:", centroid)
print(e1.get_interpolation_matrix(*centroid))

print("\nK matrix:")
print(e1.get_stiffness_matrix())
```
