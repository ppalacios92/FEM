```python
from fem.core.Node         import Node
from fem.core.Material     import Material
from fem.sections.Membrane import Membrane
from fem.validation.validation_json import FEMValidator
```

```python
# Material
Steel = Material(name='Steel', E=200000.0, nu=0.3, rho=0.0)
F1    = Membrane(name='Plate', thickness=10.0, material=Steel)
```

```python
# Output output_dir
output_dir   = r'C:\Dropbox\01. Brain\11. GitHub\FEM\examples\validation'
```

```python
from fem.elements.CST      import CST

# CST 
FEMValidator.generate(
    nodes        = [Node(0,[0.0,0.0]), Node(1,[1.0,0.0]), Node(2,[0.0,1.0])],
    element_type = CST,
    section      = F1,
    label        = 'CST_regular',
    output_dir   = output_dir,
)

FEMValidator.generate(
    nodes        = [Node(0,[0.0,0.0]), Node(1,[1.5,0.3]), Node(2,[0.2,1.2])],
    element_type = CST,
    section      = F1,
    label        = 'CST_distorted',
    output_dir   = output_dir,
)
```

```python
from fem.elements.LST      import LST

# LST 
FEMValidator.generate(
    nodes        = [Node(0,[0.0,0.0]), Node(1,[1.0,0.0]), Node(2,[0.0,1.0]),
                    Node(3,[0.5,0.0]), Node(4,[0.5,0.5]), Node(5,[0.0,0.5])],
    element_type = LST,
    section      = F1,
    label        = 'LST_regular',
    output_dir   = output_dir,
)

FEMValidator.generate(
    nodes        = [Node(0,[0.0,0.0]), Node(1,[1.5,0.3]), Node(2,[0.2,1.2]),
                    Node(3,[0.75,0.15]), Node(4,[0.85,0.75]), Node(5,[0.1,0.6])],
    element_type = LST,
    section      = F1,
    label        = 'LST_distorted',
    output_dir   = output_dir,
)
```

```python
from fem.elements.Quad4    import Quad4


# Quad4
FEMValidator.generate(
    nodes        = [Node(0,[0.0,0.0]), Node(1,[1.0,0.0]),
                    Node(2,[1.0,1.0]), Node(3,[0.0,1.0])],
    element_type = Quad4,
    section      = F1,
    label        = 'Quad4_regular',
    output_dir   = output_dir,
)

FEMValidator.generate(
    nodes        = [Node(0,[0.0,0.0]), Node(1,[1.2,0.1]),
                    Node(2,[1.1,1.3]), Node(3,[0.1,1.0])],
    element_type = Quad4,
    section      = F1,
    label        = 'Quad4_distorted',
    output_dir   = output_dir,
)
```
